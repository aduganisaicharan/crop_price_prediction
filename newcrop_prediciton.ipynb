```python
from sklearn.preprocessing import LabelEncoder
label_en = LabelEncoder()
state_arr = ['State_Bihar','State_Gujarat','State_Madhya Pradesh','State_Haryana','State_Andhra Pradesh','State_Karnataka','State_Maharashtra','State_Uttar Pradesh','State_Punjab','State_West Bengal','State_Orissa','State_Tamil Nadu','State_Rajasthan']
for i in state_arr:
    df[i] = label_en.fit_transform(df[i])
crop_arr = ['Crop_ MUSTARD','Crop_ARHAR','Crop_COTTON','Crop_GRAM','Crop_GROUNDNUT','Crop_MAIZE','Crop_MOONG','Crop_PADDY','Crop_SUGARCANE','Crop_WHEAT']
for j in crop_arr:
    df[j] = label_en.fit_transform(df[j])
# df['State_Bihar'] = label_en.fit_transform(df['State_Bihar'])
# df['State_Gujarat'] = label_en.fit_transform(df['State_Gujarat'])
# df['State_Haryana'] = label_en.fit_transform(df['State_Haryana'])
# df['State_Andhra Pradesh'] = label_en.fit_transform(df['State_Andhra Pradesh'])
# df['State_Karnataka'] = label_en.fit_transform(df['State_Karnataka'])
# df['State_Maharastra'] = label_en.fit_transform(df['State_Maharastra'])
# df['State_Uttar Pradesh'] = label_en.fit_transform(df['State_Uttar Pradesh'])
# df['State_Punjab'] = label_en.fit_transform(df['State_Punjab'])
# df['State_West Bengal'] = label_en.fit_transform(df['State_West Bengal'])
# df['State_Orissa'] = label_en.fit_transform(df['State_Orissa'])
# df['State_Tamil Nadu'] = label_en.fit_transform(df['State_Tamil Nadu'])
# df['State_Rajasthan'] = label_en.fit_transform(df['State_Rajasthan'])
# df['State_Haryana'] = label_en.fit_transform(df['State_Haryana'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
X = df.drop('Price_log',axis=1)
y = df['Price_log']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
l_model = LinearRegression()
l_model.fit(X_train,y_train)
```

```python
l_ypred = l_model.predict(X_test)
```

```python
l_mse = mean_squared_error(y_test,l_ypred)
print(l_mse)
l_mae = mean_absolute_error(y_test,l_ypred)
print(l_mae)
l_r2 = r2_score(y_test,l_ypred)
print(l_r2)
```
